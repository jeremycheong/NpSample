```python
import numpy as np

SplitLine = "\n==============================\n"

def ArrayAttr(x):
    print(f"维度数(秩)：{x.ndim}; 元素类型：{x.dtype}; 数组的维度形状：{x.shape}; 元素个数：{x.size}; 元素的字节大小：{x.itemsize}", end=SplitLine)
```

```python
x = np.arange(1, 13)
print(x, end=SplitLine)
ArrayAttr(x)
s = x[2:8:2]
print(s, end=SplitLine)
s = x[5:]
print(s, end=SplitLine)
s = x[:5]
print(s, end=SplitLine)
# 切片索引
x = np.arange(1, 65).reshape((4, 4, 4))
print(x, end=SplitLine)
ArrayAttr(x)
s = x[1:4:2]  # 只在第1个维度上起作用
print(s)
ArrayAttr(s)

s = x[:, 1:4:2]  # 只在第2个维度上起作用
print(s)
ArrayAttr(s)
s1 = x[..., 1:4:2]  # 在最后维度上起作用，省略号为自动计算的维度，且只能有一个省略号
print(s1)
ArrayAttr(s1)
s1 = x[..., 1:4:2, :]  # 在第2维度上起作用
print(s1)
ArrayAttr(s1)

s = x[:,:,1:4:2]  # 只在第3个维度上起作用
print(s)
ArrayAttr(s)

s = x[:,:2,:]
print(s)
ArrayAttr(s)
s = x[:,2:,:]
print(s)
ArrayAttr(s)
# 布尔条件选择
s = x[x > 32]
print(s)
ArrayAttr(s)
s = x[~(x > 32)]
print(s)
ArrayAttr(s)
# 花式索引，注意返回数组的维度
s = x[[0, 3]]
print(s)
ArrayAttr(s)
s = x[[0, 3], [1, 2]]
print(s)
ArrayAttr(s)
s = x[[0, 3], [1, 2], [2, 3]]  # 相当于返回点(0,1,2)和点(3，2，3)的值,这里只能使用"[]",不能使用"()"
print(s)
ArrayAttr(s)
s = x[(0, 3), (1, 2), (2, 3)]  # 相当于返回点(0,1,2)和点(3，2，3)的值
print(s)
ArrayAttr(s)

# 使用"[]"与使用"()"区别
s = x[[0, 3]]
print(s)
ArrayAttr(s)
s = x[(0, 3)]
print(s)
ArrayAttr(s)
s = x[(0, 3),]
print(s)
ArrayAttr(s)

# 不同维度使用不同的索引数组
s = x[[0, 3]][:,[1, 2]][:,:,[2, 3]]
print(s)
ArrayAttr(s)

s = x[np.ix_([0, 3], [1, 2])]
print(s)
ArrayAttr(s)
s = x[np.ix_([0, 3], [1, 2], [2, 3])]
print(s)
ArrayAttr(s)
```
