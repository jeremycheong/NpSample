```python
import numpy as np

SplitLine = "\n==============================\n"

def ArrayAttr(x):
    print(f"维度数(秩)：{x.ndim}; 元素类型：{x.dtype}; 数组的维度形状：{x.shape}; 元素个数：{x.size}; 元素的字节大小：{x.itemsize}", end=SplitLine)

def ArrayAttr(info:str, x):
    print(info + ":\n", x)
    print(f"维度数(秩)：{x.ndim}; 元素类型：{x.dtype}; 数组的维度形状：{x.shape}; 元素个数：{x.size}; 元素的字节大小：{x.itemsize}", end=SplitLine)
```

* numpy.amin() 用于计算数组中的元素沿指定轴的最小值。   
* numpy.amax() 用于计算数组中的元素沿指定轴的最大值。   
* numpy.ptp() 函数计算数组中元素最大值与最小值的差（最大值 - 最小值）。 
* numpy.percentile() 百分位数是统计中使用的度量，表示小于这个值的观察值的百分比。 函数numpy.percentile()接受以下参数。  
* numpy.median() 函数用于计算数组 a 中元素的中位数（中值）  
* numpy.mean() 函数返回数组中元素的算术平均值，如果提供了轴，则沿其计算。   
* numpy.average() 函数根据在另一个数组中给出的各自的权重计算数组中元素的加权平均值。

```python
a = np.random.randint(1, 100, size=(3, 4))
print("a:\n", a)
ArrayAttr(a)
```

```python
b_min = np.min(a)
ArrayAttr("b_min", b_min)

b_min0 = np.amin(a, axis=0)
ArrayAttr("b_min0", b_min0)
b_min1 = np.min(a, axis=1, keepdims=True)
ArrayAttr("b_min1",b_min1)
# initial: 可选参数，用于指定一个初始值，然后在数组的元素上计算最小值。
# where: 可选参数，一个布尔数组，用于指定仅考虑满足条件的元素。
b_min2 = np.min(a, axis=1, initial=20)
ArrayAttr("b_min2", b_min2)
#########################################################################
# np.amax, np.max 使用方法与np.amin, np.min相同
```

```python
ArrayAttr("a", a)

b_ptp = np.ptp(a)
ArrayAttr("b_ptp", b_ptp)

b_ptp0 = np.ptp(a, axis=0)
ArrayAttr("b_ptp0", b_ptp0)
b_ptp1 = np.ptp(a, axis=1)
ArrayAttr("b_ptp1", b_ptp1)

b_ptp2 = np.ptp(a, axis=1, keepdims=True)
ArrayAttr("b_ptp2", b_ptp2)
```

```python
ArrayAttr("a", a)

b = np.percentile(a, 100)
ArrayAttr("b", b)

# q: 要计算的百分位数，在 0 ~ 100 之间
b50 = np.percentile(a, 50)
ArrayAttr("b50", b50)

bk = np.percentile(a, 100, axis=0, keepdims=True)
ArrayAttr("bk", bk)
```

```python
ArrayAttr("a", a)

m = np.median(a)
ArrayAttr("m", m)

m0 = np.median(a, axis=0)
ArrayAttr("m0", m0)
m1 = np.median(a, axis=1, keepdims=True)
ArrayAttr("m1", m1)
```

```python
ArrayAttr("a", a)

bm = np.mean(a)
ArrayAttr("bm", bm)

bm0 = np.mean(a, axis=0)
ArrayAttr("bm0", bm0)

bm1 = np.mean(a, axis=1, keepdims=True)
ArrayAttr("bm1", bm1)
```

```python
ArrayAttr("a", a)

b_avg = np.average(a)
ArrayAttr("b_avg", b_avg)

b_avg0 = np.average(a, axis=0)
ArrayAttr("b_avg0", b_avg0)

b_avg1 = np.average(a, axis=1, keepdims=True)
ArrayAttr("b_avg1", b_avg1)

weight = np.random.randint(0, 10, size=(3, 4))
ArrayAttr("weight", weight)

# 加权平均
b_w = np.average(a, axis=0, weights=weight, keepdims=True)
ArrayAttr("b_w", b_w)
```

* np.std() 标准差是一组数据平均值分散程度的一种度量。   
标准差是方差的算术平方根。  
标准差公式如下：    
std = sqrt(mean((x - x.mean())**2))

```python
x = np.array([1, 2, 3, 4])
y = np.std(x)
ArrayAttr("y", y)

x_mean = np.mean(x)
print("x_mean: ", x_mean)

y_ = np.sqrt(np.mean((x - x_mean)**2))
print("y_: ", y_)

ArrayAttr("a", a)

b0 = np.std(a, axis=0)
ArrayAttr("b0", b0)

a_mean = np.mean(a, axis=0)
b0_ = np.sqrt(np.mean(np.power(a - a_mean, 2), axis=0))
ArrayAttr("b0_", b0_)

b1 = np.std(a, axis=1, keepdims=True)
ArrayAttr("b1", b1)
```

* np.var() 统计中的方差（样本方差）是每个样本值与全体样本值的平均数之差的平方值的平均数     
即:     
mean((x - x.mean())** 2)

```python
ArrayAttr("a", a)

bv0 = np.var(a, axis=0)
ArrayAttr("bv0", bv0)
bstd0 = np.sqrt(bv0)
ArrayAttr("bstd0", bstd0)

bv1 = np.var(a, axis=1, keepdims=True)
ArrayAttr("bv1", bv1)
```
