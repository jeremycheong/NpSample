```python
import numpy as np

SplitLine = "\n==============================\n"

def ArrayAttr(x):
    print(f"维度数(秩)：{x.ndim}; 元素类型：{x.dtype}; 数组的维度形状：{x.shape}; 元素个数：{x.size}; 元素的字节大小：{x.itemsize}", end=SplitLine)
```

```python
# 广播(Broadcast)是 numpy 对不同形状(shape)的数组进行数值计算的方式， 对数组的算术运算通常在相应的元素上进行。
a = np.arange(1, 13)
b = np.linspace(1, 12, 12)
print("a:\n", a, end=SplitLine)
print("b:\n", b, end=SplitLine)
c = a * b
print(c, end=SplitLine)
c = a.dot(b)
print(c, end=SplitLine)

# 在dim=2上的元素进行计算
a = a.reshape(3, 2, 2)
print(a)
ArrayAttr(a)
print("在dim=2上的元素进行计算", end=SplitLine)
x = np.array([2, 4])
print(x)
ArrayAttr(x)
y = a + x
print(y)
ArrayAttr(y)
y = a * x
print(y)
ArrayAttr(y)
# 效果相当于：
# x = np.tile(x, (3, 2, 1))  # 将原矩阵横向、纵向地复制
# print(x)
# ArrayAttr(x)

# 在dim=1上的元素进行计算
print("在dim=1上的元素进行计算", end=SplitLine)
# a = a.reshape(3, 2, 2)
print(a)
ArrayAttr(a)    
x = np.array([2, 4]).reshape(2, 1)
print(x)
ArrayAttr(x)
y = a + x
print(y)
ArrayAttr(y)
y = a * x
print(y)
ArrayAttr(y)
# 效果相当于：
# x = np.tile(x, (3, 1, 2))
# print(x)
# ArrayAttr(x)

# 在dim=0上的元素进行计算
print("在dim=0上的元素进行计算", end=SplitLine)
print(a)
ArrayAttr(a)
x = np.array([2, 4, 6]).reshape( 3, 1, 1)
print(x)
ArrayAttr(x)
y = a + x
print(y)
ArrayAttr(y)
y = a * x
print(y)
ArrayAttr(y)
# 效果相当于：
# x = np.tile(x, (1, 2, 2))
# print(x)
# ArrayAttr(x)
```
